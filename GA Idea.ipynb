```python
class GeneticAlgorithmMax:
    def __init__(
        self,
        population_size: int = 50,
        generations: int = 100,
        crossover_rate: float = 0.8,
        mutation_rate: float = 0.1,
        bounds: List[Tuple[float, float]] = [(-10, 10), (-10, 10), (-10, 10)],
        tournament_size: int = 3,
        elitism: int = 2,
        mutation_strength: float = 1.0,
        seed: int = 42,
    ):
        self.population_size = population_size
        self.generations = generations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        self.bounds = bounds
        self.tournament_size = tournament_size
        self.elitism = elitism
        self.mutation_strength = mutation_strength
        random.seed(seed)

    def create_individual(self) -> List[float]:
        return [
            random.uniform(low, high)
            for low, high in self.bounds
        ]

    def create_population(self) -> List[List[float]]:
        return [self.create_individual() for _ in range(self.population_size)]

    def fitness(self, individual: List[float]) -> float:
        x, y, z = individual
        return tar_funct(x, y, z)

    def tournament_selection(self, population: List[List[float]]) -> List[float]:
        candidates = random.sample(population, self.tournament_size)
        return max(candidates, key=self.fitness)

    def crossover(self, parent1, parent2):
        if random.random() > self.crossover_rate:
            return parent1[:], parent2[:]
    
        child1, child2 = [], []
        for g1, g2 in zip(parent1, parent2):
            if random.random() < 0.5:
                child1.append(g1)
                child2.append(g2)
            else:
                child1.append(g2)
                child2.append(g1)
        return child1, child2

    def mutate(self, individual: List[float]) -> List[float]:
        mutated = individual[:]
        for i, (low, high) in enumerate(self.bounds):
            if random.random() < self.mutation_rate:
                mutated[i] += random.gauss(0, self.mutation_strength)
                mutated[i] = max(low, min(high, mutated[i]))
        return mutated

    def run(self):
        population = self.create_population()

        best_individual = max(population, key=self.fitness)
        best_fitness = self.fitness(best_individual)

        history = []

        for gen in range(self.generations):
            population = sorted(population, key=self.fitness, reverse=True) #1 population!

            new_population = population[:self.elitism]

            while len(new_population) < self.population_size:
                parent1 = self.tournament_selection(population)
                parent2 = self.tournament_selection(population)

                child1, child2 = self.crossover(parent1, parent2)

                child1 = self.mutate(child1)
                child2 = self.mutate(child2)

                new_population.append(child1)
                if len(new_population) < self.population_size:
                    new_population.append(child2)

            population = new_population

            current_best = max(population, key=self.fitness)
            current_best_fit = self.fitness(current_best)

            if current_best_fit > best_fitness:
                best_fitness = current_best_fit
                best_individual = current_best[:]

            history.append(best_fitness)

            # if gen % 10 == 0 or gen == self.generations - 1:
            x, y, z = best_individual
            print(
                f"Generation {gen:3d} | "
                f"Best fitness = {best_fitness:.10f} | "
                f"x={x:.6f}, y={y:.6f}, z={z:.6f}, x-(y+z)={x-(y+z):.10f}"
            )

        return best_individual, best_fitness, history
```

```python
if __name__ == "__main__":
    ga = GeneticAlgorithmMax(
        population_size=10,
        generations=1000,
        crossover_rate=0.9,
        mutation_rate=0.2,
        bounds=[(-1000, 1000), (-1000, 1000), (-1000, 1000)],
        tournament_size=3,
        elitism=2,
        mutation_strength=0.1,
        seed=42
    )

    best_sol, best_fit, history = ga.run()

    print("\nFinal best solution:")
    print(f"x = {best_sol[0]:.6f}")
    print(f"y = {best_sol[1]:.6f}")
    print(f"z = {best_sol[2]:.6f}")
    print(f"fitness = {best_fit:.10f}")
    print(f"x - (y + z) = {best_sol[0] - (best_sol[1] + best_sol[2]):.10f}")
```
